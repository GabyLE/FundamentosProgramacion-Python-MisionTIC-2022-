## Variante 1

Dada una matriz  de nmeros enteros con tamao   , sean 0      1 la -sima fila de la matriz  y 0      1 la -sima columna de la matriz , considere las casillas
que cumplen con las siguientes dos especificaciones:

*  +  es impar (El resultado de la suma del ndice de la fila con el ndice de la columna es un nmero impar, o tambin se puede ver como las casillas intercaladas como se ve en la imagen 1).
* Las casillas que estn por encima de la diagonal secundaria (incluida la diagonal secundaria).

Para una matriz de tamao 8  8 las casillas descritas son las que estn sombreadas de
color rojo:



**Tareas:** Determine
* El valor mnimo ubicado en las casillas descritas anteriormente.
* La posicin o las posiciones donde se encuentra el valor mnimo determinado.

**Formato de entrada:**
* Matriz de numpy  de tamao   

**Restricciones:**
*  es una matriz de nmeros enteros
*    (Es un nmero natural)

**Formato de salida:**

La funcin () debe hacer los siguientes retornos en ese mismo orden:
1. Valor mnimo (Nmero entero o flotante)
2. Lista de Python con la posicin o posiciones donde se encuentra el valor mnimo, cada posicin DEBE guardarse en una tupla.



El valor mnimo que se encuentra en las casillas rojas es el nmero 5, y se encuentra en las
posiciones (0,1) y (3,0), como la posicin (1,1) NO cumple con que 1 + 1 es impar, no se
toma en cuenta.

```python
# numeros casillas impares
def solucion(a):
    lstImp = []
    for f in range(len(a)):
        for c in range(len(a[0])):
            if (f+c) % 2 != 0:
                lstImp.append(a[f][c])
                
    min = lstImp[0]
    for num in lstImp:
        if num <= min:
            min = num
            
    pos = []
    for f in range(len(a)):
        for c in range(len(a[0])):
            if (f+c)%2 != 0 and a[f][c] == min:
                pos.append((f,c))
                
    return min, pos
```

```python
matriz = [[89, 5, 23, 72],
 [51, 5, 81, 62],
 [27, 26, 88, 33],
 [ 5, 78, 11, 11]]

solucion(matriz)
```

## Variante 3

Dada una matriz  de nmeros enteros con tamao   , sean 0      1 la -sima fila de la matriz  y 0      1 la -sima columna de la matriz .

**Tareas:** Determine
* La suma de los elementos que se encuentran en la diagonal principal de la matriz 
* El producto de los elementos que se encuentran en la diagonal secundaria de la matriz 
* El mdulo entre  y , es decir %

**Formato de entrada:**
* Matriz de numpy  de tamao   

**Restricciones:**
*  es una matriz de nmeros enteros
*    (Es un nmero natural)

**Formato de salida:**
La funcin () debe hacer los siguientes retornos en ese mismo orden:
1. Suma de los elementos de la diagonal principal (Nmero entero o flotante).
2. Producto de los elementos de la diagonal secundaria (Nmero entero o flotante).
3. Mdulo entre  y  (Nmero entero o flotante).



**Ejemplo de salida:**

199

758160

169

**Explicacin:**

Los elementos de la diagonal principal son: 89, 11, 88 y 11, al sumarlos obtenemos como
resultado: 199

Los elementos de la diagonal principal son los que se encuentran en las casillas rojas:



Los elementos de la diagonal secundaria son: 72, 81, 26 y 5, al multiplicarlos obtenemos
como resultado: 758160

Los elementos de la diagonal secundaria son los que se encuentran en las casillas verdes:



Al hacer 758160%199 obtenemos 169

```python
def suma(matriz):
    sum = 0
    for f in range(len(matriz)):
        for c in range(len(matriz[0])):
            if f == c:
                sum += matriz[f][c]
    return sum

def producto(matriz):
    prod = 1
    for f in range(len(matriz)):
        for c in range(len(matriz[0])):
            if c == len(matriz) - 1 - f: 
                prod *= matriz[f][c]
    return prod

def solucion(a):
    sumaDiag = suma(a)
    productoDiag = producto(a)
    mod = productoDiag % sumaDiag
    return sumaDiag, productoDiag, mod
```

```python
matriz = [[89, 13, 23, 72],
 [29, 11, 81, 62],
 [27, 26, 88, 33],
 [ 5, 78, 11, 11]]

solucion(matriz)
```

## Variante 4

Dada una matriz  de nmeros enteros con tamao   , sean 0      1 la -sima
fila de la matriz  y 0      1 la -sima columna de la matriz .
Un nmero  es un cuadrado perfecto si    (Si  es un nmero natural).

**Tareas:** Determine
* Cantidad de cuadrados perfectos que hay en la matriz . (Si hay un mismo cuadrado perfecto que aparece ms de dos veces en la matriz , se cuenta tantas veces como aparezca en la matriz).
* La lista de los nmeros que son cuadrados perfectos en la matriz .

**Formato de entrada:**
* Matriz de numpy  de tamao   

**Restricciones:**
*  es una matriz de nmeros enteros
*    (Es un nmero natural)

**Formato de salida:**

La funcin () debe hacer los siguientes retornos en ese mismo orden:
* Cantidad de cuadrados perfectos que hay en la matriz  (Este retorno debe ser un nmero entero o flotante).
* Lista de nmeros que son cuadrados perfectos en la matriz  (Este retorno debe ser una lista de Python).



Ejemplo de salida:
3
[4, 6241, 4]

**Explicacin:**

Los nicos cuadrados perfectos que hay son 4 y 6241.

El 4 se encuentra dos veces, y el 6241 se encuentra una vez, por lo que en total hay tres
cuadrados perfectos, por eso en la lista aparece dos veces el 4 y una vez el 6241.

4 es un cuadrado perfecto, ya que 4 = 2  

6241 es un cuadrado perfecto, ya que 6241 = 79  

```python
def solucion(a):
    contar = 0
    lst = []
    
    for f in range(len(a)):
        for c in range(len(a[0])):
            num = a[f][c]
            if int(num**0.5)**2 == num:
                contar += 1
                lst.append(num)
    return contar, lst
```

```python
matriz = [[9552, 4, 5024, 6241, 2646],
 [3995, 3379, 7913, 4, 2467],
 [6313, 8128, 6249, 1949, 6405],
 [8259, 4326, 9972, 8473, 9702],
 [9149, 7726, 5972, 9577, 5686]]

solucion(matriz)
```
