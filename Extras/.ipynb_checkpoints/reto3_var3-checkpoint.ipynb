**TAREAS**
* Realizar una calculadora en Python de cadenas de bits con tres operaciones bsicas: operacin AND, operacin OR, operacin XOR, teniendo en cuenta la siguiente tabla de verdad:



**FORMATO DE ENTRADA**
La funcin (1, 2,) recibe como parmetros:
* bits1: Cadena de texto (Objeto de la clase str) que contiene una cadena de bits.
* bits2: Cadena de texto (Objeto de la clase str) que contiene una cadena de bits de la misma longitud que bits1.
* OP: Cadena de texto (Objeto de la clase str) que contiene el nombre de la operacin a realizar entre la cadena bits1 y bits2. Estas son las equivalencias:



**FORMATO DE SALIDA**
La funcin (1, 2,) debe retornar:
* resultado: Cadena de texto (Objeto de la clase str) que contiene el resultado de realizar la operacin especificada entre las cadenas bits1 y bits2 en ese orden.

**EJEMPLOS**

ENTRADA:

bits1 = "0110110110"

bits2 = "1100011101"

OP = "OR"

SALIDA:

"1110111111"

EXPLICACIN:

Las cadenas de bits recibidas en los parmetros bits1 y bits2 las podemos escribir de la siguiente manera y aplicar la tabla de verdad bit a bit:





```python
def opAND(bit1, bit2):
    if bit1 == bit2:
        return bit1
    else:
        return "0"
    
def opOR(bit1, bit2):
    if bit1 == "1" or bit2 == "1":
        return "1"
    else:
        return "0"
    
def opXOR(bit1, bit2):
    if bit1 == bit2:
        return "0"
    else:
        return "1"
    
def calculadora(bits1, bits2, OP):
    
    resultado = ""
    
    for i in range(len(bits1)):
        if OP == "AND":
            resultado += opAND(bits1[i], bits2[i])
        elif OP == "OR":
            resultado += opOR(bits1[i], bits2[i])
        else:
            resultado += opXOR(bits1[i], bits2[i])
            
    return resultado
```

```python
calculadora("0110110110", "1100011101", "OR")
```

```python
calculadora("0110110110", "1100011101", "AND")
```

```python
calculadora("0110110110", "1100011101", "XOR")
```
