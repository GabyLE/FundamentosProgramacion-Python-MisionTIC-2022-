Si no llega a este monto total, se verifica que por lo menos en el tipo de producto “Agujas” haya comprado más de 70.000.000 COP, en “Escolares” una compra superior a 30.000.000 COP y en “Hogar” compras superiores a 40.000.000 COP entonces se le asignará un recargo del 7\% a todas sus compras. Pero si solo superó los topes anteriores para algunos de los productos, se cobrará un 5\% al tipo de producto que haya superado el umbral.

Para los clientes internacionales, los precios están en dólares, se aplican reglas equivalentes a la de los clientes nacionales, pero con los siguientes umbrales y porcentaje de descuentos a aplicar: Para compras totales de 100.000 USD o superiores, un 8\% de descuento. 5\% si compra mas de 25.000 USD en Agujas, mas de 10.000 USD en Escolares y 15.000 USD en Hogar.
```python
dict = {"nombre": "César Díaz", "nacional": True, "agujas": 70000100, "escolares": 20000000, "hogar": 40000001}

def calculo_recargo(dict):
    nombre = dict["nombre"]
    nacional = dict["nacional"]
    agujas = dict["agujas"]
    escolares = dict["escolares"]
    hogar = dict["hogar"]
    
    # compra acumulada
    acum = agujas + escolares + hogar
    
    r_agujas = 0
    r_escolares = 0
    r_hogar = 0 
    
    # clientes nacionales
    if nacional:
        if acum >= 200000000:
            r_agujas = 10
            r_escolares = 10
            r_hogar = 10
            
        elif agujas > 70000 and escolares > 30000000 and hogar > 4000000:
            r_agujas = 10
            r_escolares = 10
            r_hogar = 10
            
        else:
            if agujas > 70000:
                r_agujas = 5
            if escolares > 30000000:
                r_escolares = 5
            if hogar > 4000000:
                r_hogar = 5
            
            
    # clientes internacionales
    else:
        if acum >= 100000:
            r_agujas = 8
            r_escolares = 8
            r_hogar = 8
            
        elif agujas > 25000 and escolares > 10000 and hogar > 15000:
            r_agujas = 5
            r_escolares = 5
            r_hogar = 5
        else:
            if agujas > 25000:
                r_agujas = 3
            if escolares > 10000:
                r_escolares = 3
            if hogar > 15000:
                r_hogar = 3
    
    # diccionario salida
    dSalida = {"nombre": nombre, "agujas": r_agujas, "escolares": r_escolares, "hogar": r_hogar}
    
    return dSalida

print(calculo_recargo(dict))
```

El almacén de electrodomésticos YO LE FIO ha iniciado el desarrollo de un sistema de información
para gestionar todas las ventas a crédito de los productos que tienen en exhibición. Una de las
funcionalidades principales de este sistema es realizar un módulo de simulación de crédito, en el cual a
partir de unos datos básicos de la compra como el valor del producto, la cuota inicial, número de cuotas
a financiar el saldo y el porcentaje de interés efectivo anual (%EA) que se aplicará a la venta, debe
calcular el valor de la cuota fija mensual con intereses incluidos para pagar dicho crédito en los cuotas
pactadas y los datos de cómo se amortizará la deuda, lo que incluye: número cuota, valor cuota, aporte
a capital, valor de intereses y saldo pendiente por pagar.
Para calcular la taza de interés mes vencido (TMV) que hay que aplicar al saldo pendiente por pagar
del crédito, hay que aplicar la siguiente formula:

$TMV = (1+ \% EA)^{\frac{1}{12}} - 1$

Para calcular el valor de la cuota mensual a pagar del crédito, se aplica al saldo pendiente por pagar, la
siguiente función:

$Cuota = \frac{Saldo a financiar}{\frac{1-(1+TMV)^{-meses}}{TMV}}$

donde saldo a financiar es la diferencia entre el valor del producto y la cuota inicial, y meses es el
número de cuotas a las que fue financiado saldo del producto.
Para calcular la amortización de la deuda se calcula de la siguiente forma:

$Valor intereses = Saldo a financiar * TMV$

$Capital abonado = Cuota - Valor interes$

$Nuevo Saldo = Saldo a financiar - Capital abonado$

<div class="alert alert-block alert-info">
    <b>Nota:</b> Para el último periodo, el valor del capital de esa cuota debe ser igual al saldo pendiente que se calculó para el periodo anterior </div>
